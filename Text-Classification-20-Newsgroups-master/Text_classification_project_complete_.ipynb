## For writing algo from scratch

```python
# Fit function-It will make dictionary ,dictionary with top keys as classes and having nested dictionary with keys as 
#feature nos and value as frequency of that feature in respective classes.
                
def fit(X_train, Y_train):
    result = {}
    class_values = Y_train
    for current_class in class_values:
        result[current_class] = {}
        result["total_data"] = 19997
        num_features = X_train.shape[1]
        result[current_class]["total_count"] = data[current_class-1].sum()
        for j in range(num_features ):
            result[current_class][j] = {}
            #all_possible_values = set(X_train[:, j - 1])
            #for current_value in all_possible_values:
            result[current_class][j] = data[current_class-1][j]  #it is dictionary with top keys as classes and having
                                                                 #nested dictionary with keys as feature nos and value
                                                                 #as frequency of that feature in respective classes.
                
    return result 
```

```python
# It is probability function which will calculate probability

def probability(dictionary, x, current_class):
    output=np.log(1/20)
    num_features = len(dictionary[current_class].keys()) - 1;
    for j in range(num_features ):
        if x[j]>0 :
            count_current_class_with_value_xj = dictionary[current_class][j] + 1
            count_current_class = dictionary[current_class]["total_count"] + num_features
            current_xj_probablity = np.log(count_current_class_with_value_xj)-np.log( count_current_class)
            output = output + current_xj_probablity
    return output
```

```python
# For finding best classes

def predictSinglePoint(dictionary, x):
    class_values = dictionary.keys()
    best_p = -1e100
    best_class = -1
    first_run = True
    for current_class in class_values:
        if (current_class == "total_data"):
            continue
        p_current_class = probability(dictionary, x, current_class)
        if ( first_run or p_current_class > best_p):
            best_p = p_current_class
            best_class = current_class
        first_run = False
    return best_class
```

```python
#Predict funcyion for predictions

def predict(dictionary, X_test):
    Y_predalgo = []
    for x in X_test:
        x_class = predictSinglePoint(dictionary, x)
        Y_predalgo.append(x_class)
    return Y_predalgo
```

```python
dictionary = fit(X_train,Y_train)
```

```python
Y_predalgoo = predict(dictionary,X_test)
```
